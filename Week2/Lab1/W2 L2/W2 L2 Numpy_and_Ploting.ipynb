```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = x ** 2

plt.plot(x, y)
plt.title('Line Plot: y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np

np.random.seed(42)
random_numbers = np.random.randint(1, 51, size=(3, 3))

mean_val = np.mean(random_numbers)
median_val = np.median(random_numbers)
std_val = np.std(random_numbers)
print('Random Numbers', random_numbers)
print('Mean:', mean_val)
print('Median:', median_val)
print('Standard Deviation:', std_val)
```

```python
import numpy as np

np.random.seed(42)
arr = np.random.rand(5, 5)
arr[0, 1] = np.nan
arr[2, 3] = np.nan
arr[4, 4] = np.nan

# Column means, ignoring NaN
col_means = np.nanmean(arr, axis=0)

# Find NaN positions and replace with corresponding column mean
inds = np.where(np.isnan(arr))
arr[inds] = np.take(col_means, inds[1])

print(arr)
```
